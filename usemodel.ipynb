```python
import pandas as pd

final_df = pd.read_csv('data/한화이닝별결과.csv')
otherteam_df = pd.read_csv('data/다른팀이닝별결과.csv')
```

```python
# 5회 이전용 칼럼
inning_columns_우리팀_5회까지 = ['1이닝점수_우리팀', '2이닝점수_우리팀', '3이닝점수_우리팀', '4이닝점수_우리팀', '5이닝점수_우리팀']
inning_columns_상대팀_5회까지 = ['1이닝점수_상대팀', '2이닝점수_상대팀', '3이닝점수_상대팀', '4이닝점수_상대팀', '5이닝점수_상대팀']

# '우리팀'과 '상대팀' 컬럼을 숫자형으로 변환
merged_df[inning_columns_우리팀_5회까지] = merged_df[inning_columns_우리팀_5회까지].apply(pd.to_numeric, errors='coerce')
merged_df[inning_columns_상대팀_5회까지] = merged_df[inning_columns_상대팀_5회까지].apply(pd.to_numeric, errors='coerce')

# 5회 이전 점수 합계 계산
merged_df['우리팀_5회까지'] = merged_df[inning_columns_우리팀_5회까지].sum(axis=1)
merged_df['상대팀_5회까지'] = merged_df[inning_columns_상대팀_5회까지].sum(axis=1)

# 6회 이후 칼럼
inning_columns_우리팀_6회이후 = ['6이닝점수_우리팀', '7이닝점수_우리팀', '8이닝점수_우리팀', '9이닝점수_우리팀']
inning_columns_상대팀_6회이후 = ['6이닝점수_상대팀', '7이닝점수_상대팀', '8이닝점수_상대팀', '9이닝점수_상대팀']

# '우리팀'과 '상대팀' 컬럼을 숫자형으로 변환
merged_df[inning_columns_우리팀_6회이후] = merged_df[inning_columns_우리팀_6회이후].apply(pd.to_numeric, errors='coerce')
merged_df[inning_columns_상대팀_6회이후] = merged_df[inning_columns_상대팀_6회이후].apply(pd.to_numeric, errors='coerce')

# 6회 이후 점수 합계 계산
merged_df['우리팀_6회이후'] = merged_df[inning_columns_우리팀_6회이후].sum(axis=1)
merged_df['상대팀_6회이후'] = merged_df[inning_columns_상대팀_6회이후].sum(axis=1)

# 연장 점수 칼럼
inning_columns_우리팀_연장 = ['10이닝점수_우리팀', '11이닝점수_우리팀', '12이닝점수_우리팀']
inning_columns_상대팀_연장 = ['10이닝점수_상대팀', '11이닝점수_상대팀', '12이닝점수_상대팀']

# '우리팀'과 '상대팀' 컬럼을 숫자형으로 변환
merged_df[inning_columns_우리팀_연장] = merged_df[inning_columns_우리팀_연장].apply(pd.to_numeric, errors='coerce')
merged_df[inning_columns_상대팀_연장] = merged_df[inning_columns_상대팀_연장].apply(pd.to_numeric, errors='coerce')

# 연장 점수 합계 계산
merged_df['우리팀_연장'] = merged_df[inning_columns_우리팀_연장].sum(axis=1)
merged_df['상대팀_연장'] = merged_df[inning_columns_상대팀_연장].sum(axis=1)

# 결과 확인
print(merged_df[['우리팀_5회까지', '상대팀_5회까지', '우리팀_6회이후', '상대팀_6회이후', '우리팀_연장', '상대팀_연장']])
```

```python
# NaN 값이 있는 행을 제거
merged_df = merged_df.dropna(subset=['게임결과_우리팀'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

X = merged_df[['우리팀_5회까지', '상대팀_5회까지']]
y = merged_df['게임결과_우리팀']

# 데이터 분할 (훈련 데이터와 테스트 데이터)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 랜덤 포레스트 모델 학습
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# 예측 및 평가
y_pred = model.predict(X_test)
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print(classification_report(y_test, y_pred))

# 예측 확률 계산
y_prob = model.predict_proba(X_test)

# '1' (승리) 클래스의 확률만 출력
win_prob = y_prob[:, 1]

# 예측 확률 출력
print("예측 승리 확률:")
print(win_prob)

# 승리 확률과 5회 이전 점수 시각화
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_test['우리팀_5회까지'], y=win_prob, hue=y_test, palette='coolwarm', marker='o')
plt.title('5회 이전 점수에 따른 승리 확률 예측')
plt.xlabel('우리팀 5회 이전 점수')
plt.ylabel('예측된 승리 확률')
plt.show()
```

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt

# 1. 데이터 준비
X = merged_df2[['1이닝점수_우리팀', '1이닝점수_상대팀', '2이닝점수_우리팀', '2이닝점수_상대팀',
                '3이닝점수_우리팀', '3이닝점수_상대팀', '4이닝점수_우리팀', '4이닝점수_상대팀',
                '5이닝점수_우리팀', '5이닝점수_상대팀', '6이닝점수_우리팀', '6이닝점수_상대팀',
                '7이닝점수_우리팀', '7이닝점수_상대팀', '8이닝점수_우리팀', '8이닝점수_상대팀',
                '9이닝점수_우리팀', '9이닝점수_상대팀', '10이닝점수_우리팀', '10이닝점수_상대팀',
                '11이닝점수_우리팀', '11이닝점수_상대팀', '12이닝점수_우리팀', '12이닝점수_상대팀']]
y = merged_df2['게임결과_우리팀']  # 승리 여부

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. SVM 모델 학습
svm_model = SVC(kernel='linear', random_state=42)  # SVM with linear kernel
svm_model.fit(X_train, y_train)

# 예측
y_pred = svm_model.predict(X_test)

# 정확도 및 평가 출력
accuracy = accuracy_score(y_test, y_pred)
print("SVM Accuracy:", accuracy)
print("Classification Report (SVM):")
print(classification_report(y_test, y_pred))

# 3. 특성 중요도 추출 (Permutation Importance)
result = permutation_importance(svm_model, X_test, y_test, n_repeats=10, random_state=42)

# 특성 중요도 추출
importance = result.importances_mean
features = X.columns

# 4. 특성 중요도 시각화
plt.figure(figsize=(12, 6))
plt.title("Permutation Importance - Feature Importance (SVM)")
plt.barh(features, importance, align="center")
plt.xlabel("Mean Decrease in Accuracy")
plt.show()

# 중요도 순으로 출력
importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': importance
}).sort_values(by='Importance', ascending=False)

print(importance_df)
```
